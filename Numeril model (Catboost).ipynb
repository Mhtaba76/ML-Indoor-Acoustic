```python
import pandas as pd
from sklearn.model_selection import train_test_split as tts
from catboost import CatBoostRegressor
from xgboost import XGBRFRegressor
from sklearn.multioutput import MultiOutputRegressor as mor
from sklearn.metrics import r2_score
import joblib
```

```python
df=pd.read_csv('/dataset.csv')
```

```python
df
```

```python
dfx=df.iloc[:,:14]
dfy=df.iloc[:,14:]
```

```python
dfy.columns
```

```python
dfx.columns
```

```python
g=tts(dfx,dfy,random_state=21,train_size=0.85)
```

```python
model=mor(CatBoostRegressor()).fit(g[0],g[2])
```

```python
yp=model.predict(g[1])
```

```python
r2=r2_score(g[3],yp,multioutput='raw_values')
```
